```python
# Standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

# EDA of Final Dataframe

```python
# importing data
final_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
final_2000.head()
```

```python
# importing data
final_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
final_2001.head()
```

Now I will concatenate these dataframes together to help organize and efficiently visualize the data.

```python
# Combining both dataframes into one
final_tmdb = pd.concat([final_2000, final_2001])
# Checking the first 5 rows
final_tmdb.head()
```

1. How many movies had at least some valid financial information (value > for budget OR revenue)?

```python
# Filtering budget and revenue columns
filter = (final_tmdb['budget'] > 0) & (final_tmdb['revenue'] > 0)
# Filtering the dataframe
final_tmdb = final_tmdb[filter].copy()
# Checking how many movies are left
print(f'Number of movies left: {len(final_tmdb)}')
```

```python
def format_millions(x, pos):
    """This function formats the x-axis tick labels and shortens the numbers to $M or $B"""
    if x >= 1000000000:
        return "${:.0f}B".format(x/1000000000)
    else:
        return "${:.0f}M".format(x/1000000)
```

```python
fix, ax = plt.subplots(figsize=(10, 6))
for i, col in enumerate(['budget', 'revenue']):
    sns.kdeplot(final_tmdb[col], ax=ax, label=col)
    ax.set_xlabel('Dollars')
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(format_millions))
    ax.set_title('Distribution of Budget and Revenue')
    ax.legend()

```

Judging by the graphs above, more movies receive less funding and profit than higher estimated budget and revenue of movies. There are very few movies that make it past $400M in box office revenue. Most movies that has a higher revenue and budget could be a number of Marvel movies and note worthy Avartar by James Cameron. For the most part, the average funding for a movie is around $0 to $10M for just understand 100 movies in this dataset. However, the average movie revenue tends to be around $0M to $25M with over 140 movies in this area of revenue.

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Preparing the data for the plot
plot_df = final_tmdb.apply(lambda row: row[final_tmdb['certification'].isin(['G', 'PG', 'PG-13', 'R'])])
```

```python
with plt.style.context('seaborn-muted'):
    plt.figure(figsize=(15, 10))
    sns.countplot(x='certification', data=plot_df, order=['G', 'PG', 'PG-13', 'R'], palette='Set2', zorder=3, edgecolor='black', hue='certification', dodge=False)
    plt.title('Count of Movies by MPAA Rating')
    plt.xlabel('MPAA Rating')
    plt.ylabel('Count of Movies')
    plt.yticks(np.arange(0, max(final_tmdb['certification'].value_counts()), 20))
    plt.grid(zorder=0, color='lightgray', linestyle='--', alpha=0.7, axis='y')
    plt.show()
```

Interestingly, there are more PG-13 and R MPAA rating movies than most PG and G rated films. This could indicate that more films are marketed to older audiences rather than younger viewers such as ages under the age of 13.

3. 3 & 4: What is the average revenue per certification category? What is the average budget per certification category?

```python
with plt.style.context(['seaborn-muted', 'dark_background']):
    fig, ax = plt.subplots(2, 1, figsize=(15, 10))
    for i, col in enumerate(['budget', 'revenue']):
        sns.barplot(x='certification', y=col, data=plot_df, ax=ax[i], order=['G', 'PG', 'PG-13', 'R'], palette='Set2', zorder=3, errorbar=None)
        ax[i].set_title(f'{col.capitalize()} by MPAA Rating')
        ax[i].set_xlabel('MPAA Rating')
        ax[i].set_ylabel('{} (in millions)'.format(col.capitalize()))
        ax[i].yaxis.set_major_formatter(ticker.FuncFormatter(format_millions))
        ax[i].grid(zorder=0, color='lightgray', linestyle='--', alpha=0.7, axis='y')
    plt.tight_layout()
    plt.show()
```

Dispite having productions focus on PG-13 and R rated movies, most of the movies that are G and PG seem to have higher grossing revenue. This is note worthy if a film production wanted to focus on what audiences would have a higher grossing revenue.
