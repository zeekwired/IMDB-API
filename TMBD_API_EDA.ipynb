```python
# Standard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import seaborn as sns
```

# EDA of Final Dataframe

```python
# importing data
final_2000 = pd.read_csv('Data/final_tmdb_data_2000.csv.gz')
final_2000.head()
```

```python
# importing data
final_2001 = pd.read_csv('Data/final_tmdb_data_2001.csv.gz')
final_2001.head()
```

Now I will concatenate these dataframes together to help organize and efficiently visualize the data.

```python
# Combining both dataframes into one
final_tmdb = pd.concat([final_2000, final_2001])
# Checking the first 5 rows
final_tmdb.head()
```

1. How many movies had at least some valid financial information (value > for budget OR revenue)?

```python
# Filtering budget and revenue columns
filter = final_tmdb[['budget', 'revenue']] > 0
# Filtering the dataframe
final_tmdb = final_tmdb[filter.all(axis=1)].copy()
# Checking the first 5 rows
final_tmdb.head()
```

```python
def format_millions(x, pos):
    """This function formats the x-axis tick labels and shortens the numbers to $M or $B"""
    if x >= 1000000000:
        return "${:.0f}B".format(x/1000000000)
    else:
        return "${:.0f}M".format(x/1000000)
```

```python
fig, ax = plt.subplots(2, 1, figsize=(15, 10))
for i, col in enumerate(['budget', 'revenue']):
    sns.histplot(final_tmdb[col], ax=ax[i], stat='count', zorder=3)
    ax[i].set_title(f'{col.capitalize()} Count of Movies Budget and Revenue')
    ax[i].xaxis.set_major_formatter(ticker.FuncFormatter(format_millions))
    ax[i].set_xticks(np.arange(0, max(final_tmdb[col]), 10000000))
    ax[i].set_xlabel('{} (in millions)'.format(col.capitalize()))
    ax[i].set_ylabel('Count of Movies')
    if col == 'budget':
        ax[i].set_xticks(np.arange(0, max(final_tmdb[col]), 10000000))
    else:
        ax[i].set_xticks(np.arange(0, max(final_tmdb[col]), 50000000))
    ax[i].grid(zorder=0, color='lightgray', linestyle='--', alpha=0.7, axis='y')

plt.tight_layout()
plt.show()
```

2. How many movies are there in each of the certification categories (G/PG/PG-13/R)?

```python
# Preparing the data for the plot
plot_df = final_tmdb.apply(lambda row: row[final_tmdb['certification'].isin(['G', 'PG', 'PG-13', 'R'])])
```

```python
with plt.style.context('seaborn-muted'):
    plt.figure(figsize=(15, 10))
    sns.countplot(x='certification', data=plot_df, order=['G', 'PG', 'PG-13', 'R'], palette='Set2', zorder=3)
    plt.title('Count of Movies by MPAA Rating')
    plt.xlabel('MPAA Rating')
    plt.ylabel('Count of Movies')
    plt.yticks(np.arange(0, max(final_tmdb['certification'].value_counts()), 20))
    plt.grid(zorder=0, color='lightgray', linestyle='--', alpha=0.7, axis='y')
    plt.show()
```

3. 3 & 4: What is the average revenue per certification category? What is the average budget per certification category?

```python
with plt.style.context(['seaborn-muted', 'dark_background']):
    fig, ax = plt.subplots(2, 1, figsize=(15, 10))
    for i, col in enumerate(['budget', 'revenue']):
        sns.barplot(x='certification', y=col, data=plot_df, ax=ax[i], order=['G', 'PG', 'PG-13', 'R'], palette='Set2', zorder=3, ci=None)
        ax[i].set_title(f'{col.capitalize()} by MPAA Rating')
        ax[i].set_xlabel('MPAA Rating')
        ax[i].set_ylabel('{} (in millions)'.format(col.capitalize()))
        ax[i].yaxis.set_major_formatter(ticker.FuncFormatter(format_millions))
        ax[i].grid(zorder=0, color='lightgray', linestyle='--', alpha=0.7, axis='y')
    plt.tight_layout()
    plt.show()
```
